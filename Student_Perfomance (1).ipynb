```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(clust_df)

scaled_data = scaler.transform(clust_df)
```

```python
def find_best_clusters(df, maximum_K):
    
    clusters_centers = []
    k_values = []
    
    for k in range(1, maximum_K):
        
        kmeans_model = KMeans(n_clusters = k)
        kmeans_model.fit(df)
        
        clusters_centers.append(kmeans_model.inertia_)
        k_values.append(k)
        
    
    return clusters_centers, k_values
```

```python
def generate_elbow_plot(clusters_centers, k_values):
    
    figure = plt.subplots(figsize = (12, 6))
    plt.plot(k_values, clusters_centers, 'o-', color = 'orange')
    plt.xlabel("Number of Clusters (K)")
    plt.ylabel("Cluster Inertia")
    plt.title("Elbow Plot of KMeans")
    plt.show()
```

```python
clusters_centers, k_values = find_best_clusters(scaled_data, 12)

generate_elbow_plot(clusters_centers, k_values)
```

```python
kmeans_model = KMeans(n_clusters = 4)

kmeans_model.fit(scaled_data)
```

```python
clust_df["clusters"] = kmeans_model.labels_

print(clust_df)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd


scaler = StandardScaler()
data_scaled = scaler.fit_transform(clust_df)


wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()


kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans.fit(data_scaled)


cluster_labels = kmeans.predict(data_scaled)
clust_df['cluster'] = cluster_labels
top_students = clust_df.groupby('cluster')['average_sem'].nlargest(5)
print(top_students)
```
